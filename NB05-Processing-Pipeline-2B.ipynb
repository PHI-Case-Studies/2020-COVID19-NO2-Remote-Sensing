# Sentinel 5P NO2 Data Processing Pipeline #2B

In Pipeline 2B we will:
1. Convert the downloaded NetCD4 files to Pandas DataFrames
2. Pickle the dataframes to disk

```python
from s5p_no2_tools import date_from_week, add_days, nc_to_df
```

```python
week_list = ['2019-W01','2019-W06','2019-W11','2019-W17','2020-W01','2020-W06','2020-W11','2020-W17']
working_directory = 'projects/thailand'
```

```python
import pandas as pd

gdf_list = []
for week in week_list:
    gdf_name = 'gdf-'+week+'.pickle'
    print(gdf_name)
    gdf_list.append(pd.read_pickle(working_directory+'/'+gdf_name))
```

```python
gdf_list[0]
```

```python
%%time
from os import path
for i in range(0, len(gdf_list)):
    print('========================')
    print('Processing GeoDataFrame #'+str(i+1)+'...')
    for index, row in gdf_list[i].iterrows():
        identifier = row['identifier']
        pickle_name = identifier+'.pickle'
        print('  row: '+str(index+1))
        print('  identifier:', identifier)
        if path.exists('projects/swaths/'+pickle_name):
            print('Pickle file for '+row['filename']+' exists. Skipping...')
        else:
            print('Pickle file not found.')
            df = nc_to_df('projects/swaths/'+row['filename'])
            df.to_pickle('projects/swaths/'+pickle_name)
```
